```python
#hide
import twint
import nest_asyncio
nest_asyncio.apply()


c = twint.Config()

c.Username = 'deepfates'
c.Popular_tweets = True
c.Limit = 20
c.Min_likes = 99
tweets = []
c.Store_object = True
c.Store_object_tweets_list = tweets


twint.run.Search(c)
```

```python
#hide
tweets.sort(key=lambda x: x.datetime, reverse=True)

urls = [t.link for t in tweets]
templated_links = '>twitter: ' + '\n\n>twitter: '.join(urls)

from IPython.core.getipython import get_ipython
shell = get_ipython()
```
